```python
import pandas as pd
import re

# Sample dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Email': ['alice@email.com', 'bob@work.com', 'charlie@home.com'],
    'Feature': [10, 20, 30]
}

df = pd.DataFrame(data)

# Function to mask email (show first 2 letters only)
def mask_email(email):
    return re.sub(r'(?<=.{2}).(?=.*@)', '*', email)

# Apply masking
df['Email'] = df['Email'].apply(mask_email)

print(df)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
from art.estimators.classification import KerasClassifier
from art.attacks.evasion import FastGradientMethod
import numpy as np

# 1. Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0  # Normalize to [0, 1]
x_train = np.expand_dims(x_train, axis=-1)
x_test = np.expand_dims(x_test, axis=-1)

# 2. Build a simple CNN model
model = Sequential([
    Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=1, batch_size=64) # Quick train for 1 epoch

# 3. Wrap the model for ART
# Since we normalized our data to [0, 1], our clip_values are (0.0, 1.0)
classifier = KerasClassifier(model=model, clip_values=(0.0, 1.0))

# 4. Initialize the Fast Gradient Method (FGSM) attack
attack = FastGradientMethod(estimator=classifier, eps=0.2)

# 5. Generate adversarial samples (only on a subset of 100 images to save time)
x_test_adv = attack.generate(x=x_test[:100])

# 6. Evaluate
predictions = classifier.predict(x_test_adv)
accuracy = np.sum(np.argmax(predictions, axis=1) == y_test[:100]) / 100
print(f"\nAccuracy on adversarial samples: {accuracy*100}%")
```

```python
import datetime

# 1. Global configurations / User settings
# In a real app, these would come from a database.
user_consent = True
consent_purpose = "Marketing"
data_collection_purpose = "Marketing"

def process_user_data(user_id, data, purpose, consent):
    """
    Checks if the user has consented and if the
    requested purpose matches the allowed purpose.
    """
    # Validation logic
    if not consent or purpose != data_collection_purpose:
        print(f"Error: Purpose mismatch or no consent for User {user_id}.")
        return None

    # Process only if authorized
    print(f"Processing data for {purpose}...")

    return {
        "user_id": user_id,
        "data": data,
        "timestamp": datetime.datetime.now().isoformat() # Added isoformat for readability
    }

# --- Usage ---

user_data = {"name": "Alice", "email": "alice@example.com"}

# Example of a successful run
processed = process_user_data("123", user_data, "Marketing", user_consent)

if processed:
    print("Successfully processed:", processed)
```

```python
# Raw input data simulating information received from a user or API
raw_data = {
    "username": "user1",
    "password_hash": "...",
    "email": "user@example.com",
    "location": "GPS_COORDINATES",  # Unnecessary for current use case
    "age": 25 # Necessary for age verification
}

# Function to minimize data before storage
def minimize_data(data):
    """
    Only keep necessary fields to reduce privacy risks.
    Uses a dictionary comprehension to filter keys.
    """
    allowed_fields = ["username", "email", "age"]
    minimized = {k: v for k, v in data.items() if k in allowed_fields}
    return minimized

# Execute the minimization
minimized_data = minimize_data(raw_data)

# Output the result
print(minimized_data)
# Expected Output: {'username': 'user1', 'email': 'user@example.com', 'age': 25}
```
